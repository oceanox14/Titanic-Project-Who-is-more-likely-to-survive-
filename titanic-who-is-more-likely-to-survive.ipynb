```python
training = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

**1.Understand nature of the data**

```python
training.info() # look at data types & null counts
```

```python
training.describe() # this func.provide to better understand the numeric data and the cental tendencies.
```

```python
training.describe().columns
```

```python
# look at numeric and cattegorical values seperately
df_num = training[['Age', 'SibSp', 'Parch', 'Fare']]
df_cat = training[['Survived', 'Pclass', 'Sex', 'Ticket', 'Cabin', 'Embarked']]
```

```python
# distributions for all numeric variables
import matplotlib.pyplot as plt
for i in df_num.columns:
    plt.hist(df_num[i])
    plt.title(i)
    plt.show()
```

```python
import seaborn as sns
print(df_num.corr())
sns.heatmap(df_num.corr())
```

```python
#compare survival rate across Age, SibBp, Parch & Fare
pd.pivot_table(training, index = 'Survived', values=['Age', 'SibSp', 'Parch', 'Fare'])
```

```python
#Let's explore categorical data
for i in df_cat.columns:
    sns.barplot(df_cat[i].value_counts().index,df_cat[i].value_counts()).set_title(i)
    plt.show()\
```
